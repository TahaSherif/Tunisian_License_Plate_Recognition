For the annotation files, all bounding boxes coordinates are saved in the license_plates_detection_train.csv file.  

We're going to read the .csv file and save each line in a .txt file containing the class (which is 0) and the bounding boxes coordinates. (x, y, w, h)

```
import csv

#read data from csv and save each line as an element of a list named data
with open('./license_plates_detection_train.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)  
    data.pop(0) 
print(data)
```

Let's create a directory in our colab workspace that will contain all the annotations.

```
! mkdir license_plates_detection_annotations
```

```
# importing OpenCV
import cv2
```

```
for img in data:
  yolo_format=list()
  #read image
  file_name = img[0]
  image = cv2.imread('./obj/'+file_name)
  dh = 1./image.shape[0]
  dw = 1./image.shape[1]  
  #bb center coordinates
  x = float(img[2])+((float(img[-1])-float(img[2]))/2.0)
  y = float(img[1])+((float(img[-2])-float(img[1]))/2.0)
  #height
  w = float(img[-1])-float(img[2])
  #width
  h = float(img[-2])-float(img[1])

  yolo_format.append(0)
  yolo_format.append(x*dw)
  yolo_format.append(y*dh)
  yolo_format.append(w*dw)
  yolo_format.append(h*dh)
  #saving annotation file
  annot_path = './license_plates_detection_annotations/'+file_name[0:-3]+'txt'
  with open(annot_path, "w") as filehandle:
    for elem in yolo_format:
      filehandle.write(str(elem)+' ')
```
